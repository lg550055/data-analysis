## Import libraries

```python
import numpy as np # linear algebra
import matplotlib.pyplot as plt
```

## Create chessboard class

```python
class ChessBoard:

    def __init__(self):
        black = (0.,0,0)
        board = np.full((8,8,3), black)
        white = (1.,1,1)
        board[::2, 1::2] = white
        board[1::2, ::2] = white
        self.board = board
        self.red = None, None
        self.blue = None, None

    def add_red(self, row, col):
        self.board[row,col] = (1, 0.2, 0)
        self.red = row, col

    def add_blue(self, row, col):
        self.board[row,col] = (0., 1, 1)
        self.blue = row, col

    def is_under_attack(self):
        if self.red[0] == self.blue[0] or self.red[1] == self.blue[1]:
            return True
        elif self.red in self.diagonals(self.blue[0], self.blue[1]):
            return True
        else:
            return False

    def diagonals(self, row, col):
        attackers = []
        x = row
        y = col
        while x<8 and y<8:
            x += 1
            y += 1
            attackers.append((x,y))
        x = row
        y = col
        while x>=0 and y>=0:
            x -= 1
            y -= 1
            attackers.append((x,y))
        x = row
        y = col
        while x>=0 and y<8:
            x -= 1
            y += 1
            attackers.append((x,y))
        x = row
        y = col
        while x<8 and y>=0:
            x += 1
            y -= 1
            attackers.append((x,y))
        return attackers

    def render(self):
        plt.imshow(self.board)
```

## Instantiate a chessboard and add two pieces, queens

```python
board1 = ChessBoard()
board1.add_red(1,1)
board1.add_blue(5,6)
board1.render()
```

## Ascertain if a queen is under attack under different scenarios

```python
board1.is_under_attack()
```
